2.1.2 Заполните пропущенные значения там, где это возможно.

Столбец ceiling_height

Данный столбец содержит информацию о высоте потолков. Пропуски могут быть вызваны тем, что продавец не знал эти размеры, не захотел измерять.

Посмотрим есть ли аномально высокие или низкие значения потолка

```python
print(logs.loc[(logs['ceiling_height'] < 2) | (logs['ceiling_height'] > 6), 'ceiling_height'].unique())
```

```python
logs = logs.reset_index(drop=True)
logs.loc[logs['ceiling_height'] > 22, 'ceiling_height'] = logs['ceiling_height'] / 10
logs['ceiling_height'] = logs['ceiling_height'].fillna(logs['ceiling_height'].median())
print(logs.info())
```

```python
# Заполняем все пустые ячейки 1.0, так как во всех пустых ячейках будет 1 этаж
logs['floors_total'] = logs['floors_total'].fillna(1.0)
print(logs.info())
```

В датасете содержится информация о 23699 объявлениях о продаже недвижимости.

Количество столбцов: 22

Типы данных:

Целочисленные (int64): 3 столбца

Вещественные числа (float64): 14 столбцов

Логические (bool): 2 столбца

Объектные (object): 3 столбца

Объем памяти: 3.7+ MB

```python
# В данном параметре вместо названия населенного пункта поставлено пустое место, так как не были получены данные о названии населенного пункта
logs['locality_name'] = logs['locality_name'].fillna(' ')
print(logs.info())
```

```python
# Группируем значения rooms и total_area. Далее по группам рассчитываем медиану и записываем полученное значение в определенную группу 
logs['living_area'] = logs['living_area'].fillna(logs.groupby(['rooms', 'total_area'])['living_area'].transform('median'))
print(logs.info())
```

```python
logs['balcony'] = logs['balcony'].fillna(0.0)
print(logs.info())
```

```python
# В данном случае записываем в строки, где студии значение 0.0
logs.loc[(logs['studio'] == True) & (logs['kitchen_area'].isna()), 'kitchen_area'] = 0
# В остальные с использованием медианы по площади кухни рассчитываем площадь для оставшихся строк с NaN
logs['kitchen_area'] = logs['kitchen_area'].fillna(
    logs['total_area'] * logs.groupby('rooms')['kitchen_area'].transform('median') / logs.groupby('rooms')['total_area'].transform('median'))
print(logs.info())
```

Столбец is_apartment
Имеет наибольшее количество пропусков, которое связано скорее всего с тем, что это редкий и специфический тип жилья, из-за чего при заполнии объявлений, продавцы обычных квартир не придавали значения этому атрибуту. Потому можно было бы заполнить все пропуски значениями False, но т.к. для машинного обучения рекомендуется bool переводить в числовой формат, то определим тип данных float и пропуски заполним нулями.

```python
logs['is_apartment'] = logs['is_apartment'].astype(float).fillna(0)
# Проверим количество пропусков и тип данных
print(logs['is_apartment'].info())
```

```python
# Пропуски в параметре cityCenters_nearest можно заменить на 0.0, так как не удалось получить данные о расстоянии до центра города
logs['cityCenters_nearest'] = logs['cityCenters_nearest'].fillna(0.0)
print(logs.info())
```

```python
# Пропуски в параметре parks_around3000 можно заменить на 0.0, так как в радиусе 3 км может не оказаться парков
logs['parks_around3000'] = logs['parks_around3000'].fillna(0.0)
print(logs.info())
```

```python
# Пропуски в параметре parks_nearest можно заменить на 0.0, так как не удалось получить данные о расстоянии до ближайшего парка
logs['parks_nearest'] = logs['parks_nearest'].fillna(0.0)
print(logs.info())
```

```python
# Пропуски в параметре ponds_around3000 можно заменить на 0.0, так как в радиусе 3 км нет водоемов
logs['ponds_around3000'] = logs['ponds_around3000'].fillna(0.0)
print(logs.info())
```

3. Узнайте как и какие квартиры быстрее всего продавались, а какие медленнее.

Для ответа на заданный вопрос можно составить несколько гипотез:

- Более дешевые объекты покупают быстрее. 
- Объекты на первом и последнем этажах продаются медленнее из за меньшей популярности.
- Объекты с удачным расположением продаются быстрее (расстояние от центра города и инфраструктура района (парки, водоемы)).
- Объекты с меньшей площадью могут оказаться более популярными, чем большие квартиры из за большей доступности.

Для проверки первой гипотезы проверим зависимость между last_price и days_exposition.

```python
import seaborn as sns

logs['price_thousand'] = logs['last_price'] / 1000

plt.figure(figsize=(10, 5))
sns.scatterplot(data=logs, x="price_thousand", y="days_exposition", alpha=0.5)
sns.regplot(data=logs, x="price_thousand", y="days_exposition", scatter=False, color="green") # линия тренда
plt.title("Зависимость времени продажи от цены квартиры")
plt.xlabel("Цена квартиры в тыс. руб. (price_thousand)")
plt.ylabel("Время продажи (days_exposition)")
plt.grid(True)
plt.show()
```

По этому графику можно отметить следующее:

Большинство квартир имеет цену до 10 млн рублей, и их время продажи варьируется в широком диапазоне. Вероятнее всего причина в высокой конкуренции в сегменте недорогой недвижимости.
В среднем, более дорогие квартиры продаются дольше, линия регрессии указывает на положительную корреляцию между ценой квартиры и временем продажи. 

Для проверки второй гипотезы проверим связь между floor_type (тип этажа) и days_exposition.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=logs, x="floor_type", y="days_exposition")
plt.title("Зависимость времени продажи от типа этажа")
plt.xlabel("Тип этажа (floor_type)")
plt.ylabel("Время продажи (days_exposition)")
plt.grid(True)
plt.show()
```

По этому графику можно отметить следующее:

Для всех типов этажей наблюдается большой разброс времени продажи.
Медианное время продажи для всех типов этажей примерно одинаковое, следовательно тип этажа не оказывает существенного влияния на медианное время продажи.
Фактор этажа не является решающим для времени продажи, так как первый и последний этажи не демонстрируют заметного отличия в скорости продажи по сравнению с другими этажами.

2. Создать две разные переменные: data_ohe и data_ordinal, в которые вы запишете датасет с отмасштабированными признаками из переменной data_scaler.
3. Применить технику one hot encoding к датасету в переменной data_ohe
4. Применить технику ordinal encoding к датасету в переменной data_ordinal.
5. Результаты масштабирования и кодирования признаков вывести методом head().

Пункт 1

Создание копии предобработанного датасета для обработки 

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
data_scaler = logs.copy()
```

Выделяем все численные признаки

```python
numerical_features = data_scaler.select_dtypes(include=['float64', 'int64', 'int32']).columns
```

Масштабируем числовые признаки

```python
scaler = StandardScaler()
data_scaler[numerical_features] = scaler.fit_transform(data_scaler[numerical_features])
```

Пункт 2

Создаем переменные и записываем датасет с отмасштабированными признаками

```python
data_ohe = data_scaler.copy()
data_ordinal = data_scaler.copy()
```

Пункт 3 Применить технику one hot encoding к датасету в переменной data_ohe

Выделяем категориальные признаки

```python
categorical_attributes = data_ohe.select_dtypes(include=['object', 'bool']).columns
```

Применяем One Hot Encoder

```python
one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' для избежания dummy trap
ohe_encoded = pd.DataFrame(one_hot_encoder.fit_transform(data_ohe[categorical_attributes]),
                           columns=one_hot_encoder.get_feature_names_out(categorical_attributes),
                           index=data_ohe.index)
```

Удаляем исходные категориальные признаки и добавляем закодированные

```python
data_ohe = data_ohe.drop(columns=categorical_attributes)
data_ohe = pd.concat([data_ohe, ohe_encoded], axis=1)
```

Пункт 4

Применить технику ordinal encoding к датасету в переменной data_ordinal.

```python
ordinal_encoder = OrdinalEncoder()
data_ordinal[categorical_attributes] = ordinal_encoder.fit_transform(data_ordinal[categorical_attributes])
```
